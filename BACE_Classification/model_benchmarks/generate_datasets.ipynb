```python
import pandas as pd
```

```python
bace_dataset = pd.read_csv('/home/rajeckidoyle/Documents/Classification/BACE_Classification/model_benchmarks/bace/raw/bace.csv')
bace_dataset = bace_dataset.sample(frac=1).reset_index(drop=True)
bace_dataset
```
